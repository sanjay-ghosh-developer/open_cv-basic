```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_COMPLEX
    img = cv2.putText(blank_img, text='ABCDE', org=(50,300), fontFace=font, fontScale=5, color=(255,0,0), thickness=15)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5), dtype=np.int8)
```

```python
kernel
```

```python
#Erodes an image 
result = cv2.erode(img, kernel, iterations=1)
display_img(result)
```

```python
result = cv2.erode(img, kernel, iterations=3)
display_img(result)
```

```python
img = load_img()
black_noise = np.random.randint(low=0, high=2, size=(600,600))
black_noise = black_noise * -255
black_noise_image = img + black_noise
black_noise_image[black_noise_image==-255] = 0
```

```python
display_img(black_noise_image)
```

```python
closing = cv2.morphologyEx(black_noise_image, cv2.MORPH_CLOSE, kernel)
```

```python
display_img(closing)
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_img(gradient)
```
