```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('rainbow.jpg', 0) #If we use 0 the image will be read as grayscale
plt.imshow(img, cmap='gray')
```

```python
#What thresholding does is, it converts everything below threshold value to 0 and everything above that as 255
#Here we have threshold value as 127, third parameter is the value that needs to be given to maximum value 
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
img = cv2.imread('crossword.jpg', 0)
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
#In the pic above some images are not visibe. We can make the image more clearer by incrrasing the threshold
#After increased thresold
ret, th1 = cv2.threshold(img, 180, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
#Here we are applying an adaptive threshold. Adaptive threshold means the thresholding process depends on neighbour pixels as well
#The third argument is the adaptive filter we want to use, fourth is the threshold, fifth is number of 
#pixels that need to be considered last element is the constant that need to be subtracted 
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(th2)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 8)
show_pic(th2)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 8)
show_pic(th2)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(th2)
```
